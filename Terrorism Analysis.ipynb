```python
def cleanup(sentence):
    trainList=[]
    #print(type(sentence))
    train_str=sentence.lower()
    words=words=nltk.word_tokenize(train_str)
    for word in words:
        if (not word.isdigit() and not word in stopword_set and re.search('[a-zA-Z]', word)):
            #word.translate(None, string.punctuation)
            trainList.append(word)
    train_str=' '.join(trainList)
    #print(train_str)
    return train_str
def removePunct(word):
    # define punctuation
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

    # To take input from the user
    # my_str = input("Enter a string: ")

    # remove punctuation from the string
    no_punct = ""
    for char in word:
        if char not in punctuations:
            no_punct = no_punct + char
        else:
            no_punct = no_punct + ' '
    return no_punct        
```

```python
import nltk
from nltk.corpus import stopwords
from gensim.models import Word2Vec
import string 

stopword_set = set(stopwords.words('english')+ list(string.punctuation))
sentences=[]
for i in range(len(data)):
    new_data=[]
    x=[]
    for j in range(len(data.columns)):
        sentence=data.iloc[i,j]
        sentence=str(sentence)
        if(not sentence.isdigit()):
            #if(word.isalpha() and not word in stopword_set):
            x=cleanup(sentence)
            x=removePunct(x)
            #x=' '.join(x)
            #print(x)
            new_data.append(x)
        #print(new_data)
    t=(' ').join(new_data)
    sentences.append(str(t))
print(sentences)    
```

```python
sentences = [sentence.split(' ') for sentence in sentences]
print(sentences)
model = Word2Vec(sentences, min_count=1, window=10, size=100, workers=4)
print(model)
```

```python
from gensim.models import KeyedVectors
# save model
model.save('model.bin')
print(model)
# load model
model = KeyedVectors.load('model.bin')

#outputs the average word2vec for words in this sentence
def average_vec(sentence):
    words = sentence.split()
    word_vecs = [model[w] for w in words]
    return (np.array(word_vecs).sum(axis=0)/len(word_vecs)).reshape(1,-1)
```
